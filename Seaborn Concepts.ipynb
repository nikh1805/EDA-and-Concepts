```python
df = sns.load_dataset("tips")
df
```

```python
df.dtypes
```

### tip -> dependent feature
### total_bill, sex, smoker,day, time, size -> Independent features

### Correlation with heatmap

### A correlation heatmap  uses colored cells, typically in monocromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types.

### Interrelation between feature - only be find on interger or float values
### values will be ranging from -1 to +1 (pearson coff correlation)

```python
df_numeric = df.select_dtypes(include=['float64', 'int64'])

# Now calculate the correlation matrix
correlation_matrix = df_numeric.corr()
correlation_matrix
```

```python
sns.heatmap(df_numeric.corr())
```

## Join plot

### A join plot allows to study the relationship between 2 numeric variables
### It is usually a scatterplot, a hexabin, a 2D histogram or a 2D debsity plot

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg') # Regression
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

### Distplot

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde=False, bins=10)
```

### Categorical Plots
* boxplot
* violinplot
* countplot
* bar plot

```python
sns.countplot(x='sex', data=df)  # Use x to specify the column for the x-axis
plt.show()
```

```python
sns.countplot(y='day', data=df)
```

### Bar plot

```python
sns.barplot(y='total_bill', x='sex', data=df)
```

### Box plot

```python
sns.boxplot(x='sex', y='total_bill', data=df)
plt.title('Total Bill Distribution by Sex')
plt.xlabel('Sex')
plt.ylabel('Total Bill')
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=df)
plt.title('Total Bill Distribution by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

```python
sns.boxplot(data=df, orient='v')
```

```python
sns.boxplot(data=df, x='total_bill', y='day', hue='smoker')
```

```python
sns.violinplot(x='total_bill', y='day', data=df)
```
